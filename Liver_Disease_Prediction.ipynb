```python
X = data.drop('Dataset', axis = 1)
y = data['Dataset']
```

```python
from imblearn.combine import SMOTETomek
smk = SMOTETomek(random_state=42)
X_resampled, y_resampled = smk.fit_resample(X, y)
X_resampled.shape, y_resampled.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size = 0.1, random_state = 42)
```

```python
print("Train Set: ", X_train.shape, y_train.shape)
print("Test Set: ", X_test.shape, y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=20)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test, model.predict(X_test))
```

```python
print(f"Accuracy is {round(accuracy_score(y_test, model.predict(X_test))*100,2)}")
```

```python

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

```

```python
# Feature scaling for k-NN and SVM
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier(),
    "Gradient Boosting": GradientBoostingClassifier()
}

results = []
for name, model in models.items():
    if name in ["k-NN", "SVM"]:
        model.fit(X_train_scaled, y_train)
        accuracy = model.score(X_test_scaled, y_test)
        cv_score = np.mean(cross_val_score(model, X_train_scaled, y_train, cv=5))
    else:
        model.fit(X_train, y_train)
        accuracy = model.score(X_test, y_test)
        cv_score = np.mean(cross_val_score(model, X_train, y_train, cv=5))
    
    results.append({"Model": name, "Test Accuracy": accuracy, "CV Accuracy": cv_score})

# Convert results to DataFrame
results_df = pd.DataFrame(results).sort_values(by="Test Accuracy", ascending=False)
print(results_df)
```

```python
feature_names=X.columns.tolist()
```

```python
feature_importance = pd.Series(models["Random Forest"].feature_importances_, index=feature_names)
print("Feature Importance:\n", feature_importance.sort_values(ascending=False))
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

rf = RandomForestClassifier(n_estimators=200, max_depth=20, random_state=42)
gb = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
svm = SVC(probability=True, kernel='rbf', random_state=42)

voting_clf = VotingClassifier(estimators=[('rf', rf), ('gb', gb), ('svm', svm)], voting='soft')
voting_clf.fit(X_train_scaled, y_train)
y_pred = voting_clf.predict(X_test_scaled)

accuracy = accuracy_score(y_test, y_pred)
print(f"Voting Classifier Accuracy: {accuracy:.4f}")
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(n_estimators=200, learning_rate=0.05, max_depth=10, random_state=42)
xgb.fit(X_train_scaled, y_train)
y_pred = xgb.predict(X_test_scaled)

accuracy = accuracy_score(y_test, y_pred)
print(f"XGBoost Accuracy: {accuracy:.4f}")
```

```python
from sklearn.ensemble import IsolationForest

iso = IsolationForest(contamination=0.05)  # 5% outliers
outliers = iso.fit_predict(X_resampled)
X_filtered = X_resampled[outliers == 1]
y_filtered = y_resampled[outliers == 1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_filtered, y_filtered, test_size = 0.1, random_state = 42)
```

```python
print("Train Set: ", X_train.shape, y_train.shape)
print("Test Set: ", X_test.shape, y_test.shape)
```

```python
# Feature scaling for k-NN and SVM
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(n_estimators=20),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier(),
    "Gradient Boosting": GradientBoostingClassifier()
}

results = []
for name, model in models.items():
    if name in ["k-NN", "SVM"]:
        model.fit(X_train_scaled, y_train)
        accuracy = model.score(X_test_scaled, y_test)
        cv_score = np.mean(cross_val_score(model, X_train_scaled, y_train, cv=5))
    else:
        model.fit(X_train, y_train)
        accuracy = model.score(X_test, y_test)
        cv_score = np.mean(cross_val_score(model, X_train, y_train, cv=5))
    
    results.append({"Model": name, "Test Accuracy": accuracy, "CV Accuracy": cv_score})

# Convert results to DataFrame
results_df = pd.DataFrame(results).sort_values(by="Test Accuracy", ascending=False)
print(results_df)
```

```python
rf_best = RandomForestClassifier(
    n_estimators=400,       # More trees improve stability
    max_depth=25,           # Control tree depth to prevent overfitting
    min_samples_split=4,    # Reduce overfitting
    min_samples_leaf=2,     # Make trees less deep
    max_features='sqrt',    # Select features wisely per split
    bootstrap=True,         # Better generalization
    class_weight='balanced', # Handle any class imbalance
    random_state=42
)

rf_best.fit(X_train, y_train)
y_pred = rf_best.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Further Tuned Random Forest Accuracy: {accuracy:.4f}")
```

```python
from sklearn.ensemble import VotingClassifier
from xgboost import XGBClassifier

# Define models
rf = RandomForestClassifier(n_estimators=300, max_depth=20, random_state=42)
gb = GradientBoostingClassifier(n_estimators=200, learning_rate=0.05, random_state=42)

# Ensemble Model (Soft Voting)
voting_clf = VotingClassifier(estimators=[('rf', rf), ('gb', gb)], voting='soft')
voting_clf.fit(X_train_scaled, y_train)

y_pred = voting_clf.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
print(f"Voting Classifier Accuracy: {accuracy:.4f}")
```

```python
xgb = XGBClassifier(n_estimators=300, learning_rate=0.05, max_depth=10, random_state=42)
xgb.fit(X_train_scaled, y_train)

y_pred = xgb.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
print(f"XGBoost Accuracy: {accuracy:.4f}")
```
