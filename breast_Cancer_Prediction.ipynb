```python
drp = [
    "compactness_se",
    "fractal_dimension_worst",
    "texture_se",
    "fractal_dimension_se",
    "smoothness_se",
    "smoothness_mean",
    "concave points_se",
    "fractal_dimension_mean",
    "concavity_se",
    "symmetry_se",
    "symmetry_mean",
    "compactness_mean",
    "Unnamed: 32"
]
```

```python
X_train_new = X_train.drop(drp, axis=1)
X_test_new = X_test.drop(drp, axis=1)
```

```python
X_test_new.columns
```

```python
classifier2 = RandomForestClassifier(
    n_jobs=-1,               # Uses all available CPU cores for parallel processing
    n_estimators=200,        # Number of trees in the forest
    bootstrap=True,          # Enables bootstrapping for randomness
    criterion='gini',        # Gini impurity for node splitting
    max_depth=20,            # Maximum depth of each tree
    max_features=6,          # Recommended: sqrt(12) ≈ 3.46 (rounded to 6 for better diversity)
    min_samples_leaf=1       # Minimum samples required in a leaf node
)
```

```python
classifier2.fit(X_train_new, y_train)
```

```python
confusion_matrix(y_test, classifier2.predict(X_test_new))
```

```python
print(f"Accuracy is {round(accuracy_score(y_test, classifier2.predict(X_test_new))*100,2)}%")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score, StratifiedKFold
import pandas as pd

# Define classifiers with optimal hyperparameters
models = {
    "Logistic Regression": LogisticRegression(max_iter=5000),
    "Support Vector Machine": SVC(kernel='rbf', probability=True, C=1.0),
    "K-Nearest Neighbors": KNeighborsClassifier(n_neighbors=5),
    "Decision Tree": DecisionTreeClassifier(max_depth=10, criterion="gini"),
    "Random Forest": RandomForestClassifier(n_estimators=200, max_depth=20, random_state=42)
}

final_scores = {}

# Stratified K-Fold for better accuracy estimation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for name, model in models.items():
    # Cross-validation to get a more stable accuracy
    cv_scores = cross_val_score(model, X_train_new, y_train, cv=cv, scoring="accuracy")
    model.fit(X_train_new, y_train)
    predictions = model.predict(X_test_new)
    test_accuracy = accuracy_score(y_test, predictions)

    final_scores[name] = {"CV Accuracy": cv_scores.mean(), "Test Accuracy": test_accuracy}

    print(f"{name}: CV Accuracy = {cv_scores.mean():.4f}, Test Accuracy = {test_accuracy:.4f}")

# Convert accuracy scores into a DataFrame
accuracy_df = pd.DataFrame.from_dict(final_scores, orient="index").reset_index()
accuracy_df.columns = ["Classifier", "CV Accuracy", "Test Accuracy"]

print("\nFinal Accuracy Results:")
print(accuracy_df)
```
